# Creating a Phylogenetic Tree Using Biopython and Mathplotlib

```python
# import section of the script, all the required modules are loaded here
from Bio.Phylo.TreeConstruction import DistanceCalculator
from Bio import Phylo, AlignIO, SeqIO, Seq
import matplotlib.pyplot as plt
```

# Creating path to the species.aln file


```python
with open("./cox1_sequences/species.aln", "r") as aln:
    alignment = AlignIO.read(aln, "clustal")
    print(type(alignment))
```

# Biopython Distance Calculator
https://biopython.org/docs/1.75/api/Bio.Phylo.TreeConstruction.html

```python
calculator = DistanceCalculator("identity")
```

```python
distance_matrix = calculator.get_distance(alignment)
print(distance_matrix)
```

```python
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
constructor = DistanceTreeConstructor(calculator)
```

```python
species_tree = constructor.build_tree(alignment)
species_tree.rooted = True
print(species_tree)
```

# Calculation Script
# Calculating match between two sequences

```python
def calcSeq(seqA, seqB):
    numPlaces = min(len(seqA), len(seqB))
    score = 0.0
    for i in range(numPlaces):
        if seqA[i] == seqB[i]:
            score += 1.0
    return 100.0 * score/numPlaces
    #round(score/numPlaces,2)
```

```python
# Observe the match between homo_sapiens and pan_paniscus is 90%
print(calcSeq(hylobates_lar,drosophila_melanogaster))
```

# Using the variables created above for the following script
# Will be counting their COX1 GENE Nucleotide Frequencies 

```python
def frequencyCount(Seq):
    #Temporary frequency dictionary
    tmpFreqDict = {"A": 0, "T": 0, "C": 0, "G":0, "N":0}
    #Counting process
    for nucleotide in Seq:
        tmpFreqDict[nucleotide] += 1
    return tmpFreqDict
```
